```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# 1. Initialiser l'analyseur (une seule fois)
analyzer = SentimentIntensityAnalyzer()

print("--- Test VADER (Analyse de Sentiment) ---")
print("L'analyseur est prêt.")

# ---
# Colle ton string à tester ici :
# ---
texte_a_tester = "This product is okay"

print(f"\nANALYSE DU TEXTE : '{texte_a_tester}'")

# 2. Lancer l'analyse
scores = analyzer.polarity_scores(texte_a_tester)

print("\n--- SCORES DÉTAILLÉS ---")
print(scores)

# 3. Logique de décision (celle de votre futur moteur)
compound_score = scores['compound']
sentiment_name = "Neutre" # Défaut

if compound_score >= 0.05:
    sentiment_name = 'Positif'
elif compound_score <= -0.05:
    sentiment_name = 'Négatif'
    
print("\n--- RÉSULTAT FINAL ---")
print(f"-> Sentiment: {sentiment_name}")
print(f"-> Score (Compound): {compound_score}")
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# --- Les 4 Posts à Tester ---

post_1 = """
LG 55” Class C5 Series OLED evo AI 4K UHD Smart webOS TV (2025) 
Best Buy has it for $1396.99 Haven’t purchased a new TV for ~7 years and just trying to make a good decision. Thanks! 
"""

post_2 = """
I'm really interested in a Phillips TV because of the ambilight, as a friend had one and I really love light shows. 
I didn¨t really think much of the image quality though, as I was too dazzled by the lights, and I'm also not friends with him any longer. 
I'm looking at a model called Philips 65" 4K LED Google TV 65PUS8100/12 Does anyone know how their TVs are? 
"""

post_3 = """
I have a Philips 55OLED706 TV with three Philips Play lamps under the TV and a bedside lamp with an E17 bulb. 
After replacing the bridge with the PRO version, the TV doesn't recognize the new bridge and I can't connect the TV to a full Ambilight system. 
Has anyone else experienced a similar problem? Is there a solution? 
"""

post_4 = """
As stated in the title I noticed in the last few days (was 2 weeks away before), that my hue lights stay synched with the Philips TV 75PML9506 even when the TV is already off. 
I am using ambilight + hue (no sync box) lounge mode is off, CEC settings activated (sonos must turn on and off as well). No quick start option available on the TV. 
In the hue app I tried deleting all entertainment areas + changing the desync behaviour value to ‚turn off lights‘ when the TV creates a new area after turning the TV on again. 
Internet restarted, hue box restarted, deleted hue box in the app + in apple homekit and also deleted the hue app, made a fresh setup for everything (expect disconnecting all lights from the bridge). 
On the TV I made a fresh setup for ambilights as well. Also plugged out power cable from the TV after shutting it off and checked app, even then it‘s shoving ‚synched‘… 
It‘s very annoying I can‘t turn the lights off with siri or homekit before going into the hue app and stopping the sync manually. 
Before it was working as expected, it desynched automatically after turning the TV off. Does anyone have the same issue or any solution? 
"""

posts_to_test = [post_1, post_2, post_3, post_4]

# --- Initialisation de VADER ---
analyzer = SentimentIntensityAnalyzer()

print("--- DÉBUT DU TEST VADER ---")

# --- Boucle d'analyse ---
for i, post_text in enumerate(posts_to_test):
    
    print(f"\n================ POST {i+1} ================")
    print(f"TEXTE: {post_text[:150]}...") # Affiche un aperçu
    
    # 1. Analyser
    scores = analyzer.polarity_scores(post_text)
    compound_score = scores['compound']
    
    # 2. Décider
    sentiment_name = "Neutre"
    if compound_score >= 0.05:
        sentiment_name = 'Positif'
    elif compound_score <= -0.05:
        sentiment_name = 'Négatif'
        
    # 3. Afficher
    print(f"\n--- RÉSULTAT (POST {i+1}) ---")
    print(f"   -> Score Compound: {compound_score}")
    print(f"   -> Verdict Final: {sentiment_name}")
    print(f"   -> Scores Détaillés: {scores}")

print("\n--- FIN DU TEST ---")
```
