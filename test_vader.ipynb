```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# 1. Initialiser l'analyseur (une seule fois)
analyzer = SentimentIntensityAnalyzer()

print("--- Test VADER (Analyse de Sentiment) ---")
print("L'analyseur est prêt.")

# ---
# Colle ton string à tester ici :
# ---
texte_a_tester = "This product is okay"

print(f"\nANALYSE DU TEXTE : '{texte_a_tester}'")

# 2. Lancer l'analyse
scores = analyzer.polarity_scores(texte_a_tester)

print("\n--- SCORES DÉTAILLÉS ---")
print(scores)

# 3. Logique de décision (celle de votre futur moteur)
compound_score = scores['compound']
sentiment_name = "Neutre" # Défaut

if compound_score >= 0.05:
    sentiment_name = 'Positif'
elif compound_score <= -0.05:
    sentiment_name = 'Négatif'
    
print("\n--- RÉSULTAT FINAL ---")
print(f"-> Sentiment: {sentiment_name}")
print(f"-> Score (Compound): {compound_score}")
```
